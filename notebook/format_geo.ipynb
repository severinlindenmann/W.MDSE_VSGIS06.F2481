```python
import geopandas as gpd
import pandas as pd

# Define the paths to each shapefile
file_paths = [
    'seen_poly.shp',
]

# Load each shapefile, then export it to a newline-delimited GeoJSON file
for file_path in file_paths:
    # Load the shapefile into a GeoDataFrame
    gdf = gpd.read_file(file_path)
    
    # Define the name of the output GeoJSON file based on the original shapefile name
    # Replace the file extension from .shp to .geojsonl for newline-delimited GeoJSON
    base_name, _ = os.path.splitext(file_path)
    output_file_path = base_name + '.geojsonl'
    
    # Export the GeoDataFrame to a newline-delimited GeoJSON file
    gdf.to_file(output_file_path, driver='GeoJSONSeq')
    
    print(f"Exported {file_path} to {output_file_path}")
```
